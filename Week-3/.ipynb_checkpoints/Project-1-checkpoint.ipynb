```python
import tkinter as tk
from tkinter import messagebox
import csv

# Function to check employee
def check_employee():
    name = name_entry.get().strip()
    department = dept_entry.get().strip()

    found = False
    department_members = []

    try:
        with open("GIG-logistics.csv", "r") as file:
            reader = csv.DictReader(file)

            for row in reader:
                if row["Department"] == department:
                    department_members.append(row["Name"])

                if row["Name"] == name and row["Department"] == department:
                    found = True

        if found:
            # Remove the user from list so only "other members" show
            if name in department_members:
                department_members.remove(name)

            members_text = "\n".join(department_members) if department_members else "No other members"

            messagebox.showinfo(
                "Success",
                f"Welcome {name}!\n\nOther members in {department}:\n{members_text}"
            )
        else:
            messagebox.showerror(
                "Error",
                "Employee does not exist"
            )

    except FileNotFoundError:
        messagebox.showerror("Error", "CSV file not found!")


# Create main window
root = tk.Tk()
root.title("GIG Logistics Employee Verification")
root.geometry("400x250")

# Name label and entry
name_label = tk.Label(root, text="Enter Name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

# Department label and entry
dept_label = tk.Label(root, text="Enter Department:")
dept_label.pack()
dept_entry = tk.Entry(root)
dept_entry.pack()

# Submit button
submit_button = tk.Button(root, text="Check Employee", command=check_employee)
submit_button.pack(pady=10)

# Run app
root.mainloop()
```
