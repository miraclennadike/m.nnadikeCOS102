```python
import tkinter as tk
from tkinter import ttk
```

```python
class Zenith:
    def unique_services(self):
        pass
    
    def mutual_services(self):
        return ["Lines of credit", "Investment management", "Insurance"]
```

```python
class RetailBanking(Zenith):
    def unique_services(self):
        return [ "Retirement and education accounts","Loans and mortgages","Checking and saving"]
```

```python
class GlobalBanking(Zenith):
    def unique_services(self):
        return ["Multi-currency services","Foreign currency accounts","Foreign currency credit cards","Transborder advisory services","Liquidity management"]
```

```python
class CommercialBanking(Zenith):
    def unique_services(self):
        return ["Advisory services"]
```

```python
root = tk.Tk()
root.title("Zenith Bank Services")
root.geometry("400x400")
```

```python
# Name label and entry
tk.Label(root, text="Employee Name:").pack()
name_entry = tk.Entry(root)
name_entry.pack()

# Division dropdown
tk.Label(root, text="Select Division:").pack()
division = ttk.Combobox(root, values=[
    "Retail", "Global", "Commercial"
])
division.pack()
```

```python
result = tk.Text(root, height=10, width=40)
result.pack()
```

```python
def show_services():
    result.delete("1.0", tk.END)
    
    name = name_entry.get()
    div = division.get()
    
    if div == "Retail":
        obj = RetailBanking()
    elif div == "Global":
        obj = GlobalBanking()
    elif div == "Commercial":
        obj = CommercialBanking()
    else:
        result.insert(tk.END, "Please select a division")
        return
    
    services = obj.unique_services() + obj.mutual_services()
    
    result.insert(tk.END, f"Employee: {name}\n")
    result.insert(tk.END, f"Division: {div}\n\n")
    
    for service in services:
        result.insert(tk.END, f"- {service}\n")
```

```python
tk.Button(root, text="Show Services", command=show_services).pack()
```

```python
root.mainloop()
```
