```python
import pandas as pd

df = pd.read_csv("AB_NYC_2019.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
df = df[df['price'] > 0]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['price'], bins=50)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

VISUALIZATION 1 ( INSIGHT )
1) Most listings are low-price
2) There are few expensive listings

```python
df['room_type'].value_counts().plot(kind='bar')
plt.title("Room Type Distribution")
plt.show()
```

VISUALIZATION 2 ( INSIGHT )
1) The most common is Entire home/apartment
2) The least common is Shared room

```python
plt.figure()
df.groupby('neighbourhood_group')['price'].mean().plot(kind='line', marker='o')
plt.title("Average Price by Neighbourhood Group")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Price")
plt.show()
```

VISUALIZATION 3 ( INSIGHT )
1) Manhattan has the highest prices
2) Bronx has the lowest prices
