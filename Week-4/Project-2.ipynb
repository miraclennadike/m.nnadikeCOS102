```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df = df.dropna()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df = df.drop(columns=['Cabin'])
```

```python
df = df.drop_duplicates()
```

```python
plt.figure()
df['Survived'].value_counts().plot(kind='bar')
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

VISUALIZATION 1 ( INSIGHT )
1) More passengers died than survived

```python
plt.figure()
data = df.groupby('Sex')['Survived'].mean()
plt.plot(data.index, data.values, marker='o')
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.xticks(data.index)
plt.show()
```

VISUALIZATION 2 ( INSIGHT )
1) Females have a significantly higher survival rate than males

```python
plt.figure()
plt.hist(df['Age'], bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

VISUALIZATION 3 ( INSIGHT )
1) More passengers were young adults
