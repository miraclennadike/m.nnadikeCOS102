```python
import numpy as np

print("Polynomial Root Finder")
print("1. Quadratic Equation (Ax^2 + Bx + C = 0)")
print("2. Cubic Equation (Ax^3 + Bx^2 + Cx + D = 0)")
print("3. Quartic Equation (Ax^4 + Bx^3 + Cx^2 + Dx + E = 0)")

choice = int(input("Choose an option (1-3): "))

# Quadratic Equation
if choice == 1:
    print("\nQuadratic Equation")

    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))

    coeff = [A, B, C]
    roots = np.roots(coeff)

    print("Roots are:", roots)

# Cubic Equation
elif choice == 2:
    print("\nCubic Equation")

    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    D = float(input("Enter D: "))

    coeff = [A, B, C, D]
    roots = np.roots(coeff)

    print("Roots are:", roots)

# Quartic Equation
elif choice == 3:
    print("\nQuartic Equation")

    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    D = float(input("Enter D: "))
    E = float(input("Enter E: "))

    coeff = [A, B, C, D, E]
    roots = np.roots(coeff)

    print("Roots are:", roots)

else:
    print("Invalid choice")
```
