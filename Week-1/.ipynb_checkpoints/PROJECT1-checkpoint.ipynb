```python
# SIMPLE INTEREST
P = 1000
R = 5
T = 2

A_simple = P * (1 + (R/100) * T)

print("Simple Interest Amount =", A_simple)


# COMPOUND INTEREST
P = 1000
R = 5
n = 1
t = 2

A_compound = P * (1 + (R/(100*n))) ** (n*t)

print("Compound Interest Amount =", A_compound)


# ANNUITY PLAN
PMT = 200
R = 5
n = 1
t = 2

A_annuity = PMT * (((1 + (R/(100*n))) ** (n*t) - 1) / (R/(100*n)))

print("Annuity Amount =", A_annuity)
```
