```python
import random
import tkinter as tk
from tkinter import messagebox
```

```python
employees = ["Mary Evans", "Eyo Ishan", "Durojaye Dare", "Adams Ali",
             "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James",
             "Michell Taiwo", "Abraham Jones", "Nicole Anide",
             "Kosi Korso", "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"]

tasks = ["Loading", "Transporting", "Reviewing Orders", "Customer Service", "Delivering Items"]
```

```python
class Employee:
    
    def __init__(self, name):
        self.name = name

    def check_employee(self):
        return self.name in employees

    def take_attendance(self):
        return f"Attendance taken for {self.name}"

    def assign_task(self):
        task = random.choice(tasks)
        return f"Task assigned: {task}"

    def refuse_access(self):
        return "Access Denied! You are not an employee."
```

```python
def submit():
    name = entry.get()
    emp = Employee(name)

    if emp.check_employee():
        attendance = emp.take_attendance()
        task = emp.assign_task()

        messagebox.showinfo("Success", f"{attendance}\n{task}")
    else:
        messagebox.showerror("Error", emp.refuse_access())
```

```python
root = tk.Tk()
root.title("Employee Attendance System")
root.geometry("400x200")
```

```python
label = tk.Label(root, text="Enter your name:")
label.pack(pady=10)

entry = tk.Entry(root)
entry.pack(pady=10)

button = tk.Button(root, text="Submit", command=submit)
button.pack(pady=10)
```

```python
root.mainloop()
```
