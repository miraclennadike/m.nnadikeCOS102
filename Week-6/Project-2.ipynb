```python
import tkinter as tk
from tkinter import messagebox
```

```python
class Delivery:
    
    def __init__(self, location, weight):
        self.location = location
        self.weight = weight

    def calculate_price(self):
        if self.location == "PAU":
            if self.weight >= 10:
                return 2000
            else:
                return 1500

        elif self.location == "Epe":
            if self.weight >= 10:
                return 5000
            else:
                return 4000

        else:
            return None
```

```python
def calculate():
    location = entry_location.get()
    
    try:
        weight = float(entry_weight.get())
    except:
        messagebox.showerror("Error", "Enter a valid weight")
        return

    delivery = Delivery(location, weight)
    price = delivery.calculate_price()

    if price is not None:
        messagebox.showinfo("Price", f"Delivery cost is ₦{price}")
    else:
        messagebox.showerror("Error", "Invalid location (Use PAU or Epe)")
```

```python
root = tk.Tk()
root.title("Delivery Price Calculator")
root.geometry("400x250")
```

```python
label1 = tk.Label(root, text="Enter Location (PAU or Epe):")
label1.pack(pady=5)

entry_location = tk.Entry(root)
entry_location.pack(pady=5)

label2 = tk.Label(root, text="Enter Weight (kg):")
label2.pack(pady=5)

entry_weight = tk.Entry(root)
entry_weight.pack(pady=5)
```

```python
button = tk.Button(root, text="Calculate Price", command=calculate)
button.pack(pady=20)
```

```python
root.mainloop()
```
